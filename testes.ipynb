## Testes do projeto de PDF

```python
projeto = input('Digite a descrição do projeto: ')
horas_previstas = int(input('Digite a quantidade de horas previstas: '))
valor_hora = int(input('Digite o valor da hora trabalhada (R$): '))
prazo = input('Digite o prazo estimado: ')
```

```python
print(projeto)
print(horas_previstas)
print(valor_hora)
print(prazo)
```

#### Calcular o valor das horas previstas x o valor da horas

```python
valor_projeto = horas_previstas * valor_hora
```

```python
print(valor_projeto)
```

### Gerando PDF do orcamento

```python
from fpdf import FPDF # type: ignore
```

```python
pdf = FPDF()

pdf.add_page()
pdf.set_font('Arial')
pdf.image("template.png", x=0, y=0)

pdf.text(115, 145, projeto)
pdf.text(115, 160, str(horas_previstas))
pdf.text(115, 175, str(valor_hora))
pdf.text(115, 190, prazo)
pdf.text(115, 205, str(valor_projeto))


pdf.output('Orçamento.pdf')
print("\033[1;32mOrçamento gerado com sucesso!\033[0m")
```
